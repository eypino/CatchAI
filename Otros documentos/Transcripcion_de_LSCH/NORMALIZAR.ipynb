```python
import re

def limpiar_categoria_avanzada(cat):
    if pd.isna(cat):
        return None

    c = str(cat).lower().strip()

    # Quitar prefijos comunes de error
    c = re.sub(r'^(esp[a-z]*\.?:?)', '', c)       # elimina 'esp.', 'español:', etc.
    c = re.sub(r'^[\.\s]+', '', c)                # puntos o espacios iniciales
    c = re.sub(r'\s+', ' ', c)                    # normaliza espacios

    # Si tiene una palabra al final que no pertenece (como "mancha", "blanco"), elimínala
    c = re.sub(r'\b(mancha|blanco|flojo|negro|rojo|verde|azul)\b.*', '', c)

    # --- Buscar patrones típicos de categorías ---
    # Sustantivos
    if re.search(r'\bsust', c):
        if 'm' in c and 'f' in c:
            return "sust. m. y f."
        elif re.search(r'\bm', c):
            return "sust. m."
        elif re.search(r'\bf', c):
            return "sust. f."
        elif re.search(r'\bp', c):
            return "sust. p."
        else:
            return "sust."

    # Verbos
    if re.search(r'\bv', c):
        if 'intr' in c:
            return "v. intr."
        if 'tr' in c:
            return "v. tr."
        return "v."

    # Adjetivos
    if re.search(r'\badj', c):
        return "adj."

    # Adverbios
    if re.search(r'\badv', c):
        return "adv."

    # Interjecciones
    if re.search(r'\bint', c) or 'interj' in c:
        return "interj."

    # Locuciones
    if re.search(r'\bloc', c):
        return "loc."

    # Pronombres
    if re.search(r'\bpron', c):
        return "pron."

    return "otro"


# Aplicar la limpieza mejorada
df["Categoria_limpia"] = df["Categoria"].apply(limpiar_categoria_avanzada)

# Ver el nuevo resumen
df["Categoria_limpia"].value_counts()
```

```python
import re
import numpy as np

# decide la columna objetivo (modificamos in-place)
target_col = "Categoria_limpia" if "Categoria_limpia" in df.columns else "Categoria"

# patrones (regex) -> etiqueta canónica que quieres que quede
patrones_map = {
    r'\binterj\b(?:\.|$)': 'interj.',
    r'\bint\b(?:\.|$)': 'interj.',
    r'\bloca?t\b(?:\.|$)': 'loc.',      # loc o loca? para robustez
    r'\bloc\b(?:\.|$)': 'loc.',
    r'\bprep\b(?:\.|$)': 'prep.',
    r'\bconj\b(?:\.|$)': 'conj.',
    r'\bpron\b(?:\.|$)': 'pron.',
    r'\badv\b(?:\.|$)': 'adv.',
    r'\badj\b(?:\.|$)': 'adj.',
    r'\bsust(?:\.|\s)*m(?:\.|\s)*(?:\s*y\s*f|\s*o\s*f)?': 'sust. m. y f.',  # capturamos m y f si aparecen juntos
    r'\bsust(?:\.|\s)*m\b': 'sust. m.',
    r'\bsust(?:\.|\s)*f\b': 'sust. f.',
    r'\bsust(?:\.|\s)*p\b': 'sust. p.',
    r'\bsust\b(?:\.|$)': 'sust.',
    r'v(?:\.|\s)*tr(?:\.|$)': 'v. tr.',
    r'v(?:\.|\s)*intr?(?:\.|$)': 'v. intr.',
    # variantes explícitas con "v. tr." con espacios, puntos o sin ellos
    r'v\.*\s*tr\b': 'v. tr.',
    r'v\.*\s*intr\b': 'v. intr.',
    # casos raros escritos en castellano
    r'\bespañol(?:\.|:)?\s*sust': 'sust.',
}

# Compila las regex para velocidad
patrones_compilados = [(re.compile(pat, flags=re.IGNORECASE), etiqueta) for pat, etiqueta in patrones_map.items()]

def extraer_y_reemplazar_categoria(row):
    sinon = row.get("Sinonimos")
    if pd.isna(sinon):
        return row  # nada que hacer

    s = str(sinon)
    s_original = s

    found_label = None
    found_span = None

    # buscamos la primera etiqueta que aparezca en Sinonimos
    for regex, etiqueta in patrones_compilados:
        m = regex.search(s)
        if m:
            found_label = etiqueta
            found_span = m.span()
            break

    if found_label:
        # 1) Reemplazar la columna de categoria (siempre la sobrescribimos)
        row[target_col] = found_label

        # 2) Eliminar la porción encontrada del texto de Sinonimos
        # eliminar la coincidencia concreta (primera ocurrencia)
        s = regex.sub('', s, count=1)

        # 3) Limpiar restos comunes (puntos, comas sobrantes, paréntesis, etc.)
        s = re.sub(r'^[\s\.,;:\-–—]+', '', s)   # limpieza al inicio
        s = re.sub(r'[\s\.,;:\-–—]+$', '', s)   # limpieza al final
        s = re.sub(r'\s+', ' ', s).strip()

        # 4) Si el texto restante queda vacío o solo símbolos, poner NaN
        if s == "" or re.fullmatch(r'[\W_]+', s):
            s = np.nan

        row["Sinonimos"] = s
    else:
        # Si no encontramos etiqueta, no tocamos Categoria; solo limpiamos espacios extremos
        s = re.sub(r'^\s+|\s+$', '', s)
        row["Sinonimos"] = s if s else np.nan

    return row

# Aplicar a todo el df (modifica el df in-place, reasignamos para ser explícitos)
df = df.apply(extraer_y_reemplazar_categoria, axis=1)

# Verifica ejemplos (los índices que pusiste antes)
indices = [407, 498, 590, 642, 967, 976, 1010, 1372, 1412, 1494, 1544, 1551, 1552, 1599, 1641, 1650]
indices_existentes = [i for i in indices if i in df.index]
df.loc[indices_existentes, ["Palabra", target_col, "Sinonimos"]]
```

```python
import re

# Detectar si el texto tiene frases largas (más de 3 palabras seguidas o presencia de punto final)
def es_parrafo_o_frase(texto):
    if pd.isna(texto):
        return False
    t = str(texto).strip()

    # Si tiene punto, salto de línea o más de 3 palabras → marcar como párrafo o frase
    if "." in t or "\n" in t:
        return True
    palabras = re.findall(r'\b\w+\b', t)
    return len(palabras) > 3  # más de 3 palabras se considera largo

# Crear una columna temporal solo para revisar (puedes borrarla después)
df["Tiene_parrafo_sinonimos"] = df["Sinonimos"].apply(es_parrafo_o_frase)

# Filtrar los casos sospechosos
parrafos = df[df["Tiene_parrafo_sinonimos"]]

# Mostrar ejemplos
parrafos[["Palabra", "Sinonimos"]].head(50)
```

```python
import re

def es_parrafo_o_frase_mejorado(texto):
    if pd.isna(texto):
        return False

    t = str(texto).strip()

    # Palabras totales
    palabras = re.findall(r'\b\w+\b', t)
    num_palabras = len(palabras)

    # Contar puntos
    num_puntos = t.count(".")

    # Si hay más de 4 palabras → sospechoso
    if num_palabras > 4:
        return True

    # Si hay más de 1 punto → sospechoso
    if num_puntos > 1:
        return True

    # Si hay punto en medio del texto (no solo al final)
    if re.search(r'\.\s+\w', t):  # punto seguido de palabra
        return True

    # Si contiene conectores de frase (indicativo de oración)
    if re.search(r'\b(que|para|cuando|como|porque|donde|por|una|con)\b', t.lower()):
        return True

    # Si solo tiene un punto al final o no tiene punto → se considera válido
    return False


# Aplicar detector
df["Tiene_parrafo_sinonimos"] = df["Sinonimos"].apply(es_parrafo_o_frase_mejorado)

# Filtrar solo los sospechosos
parrafos = df[df["Tiene_parrafo_sinonimos"]]

print(f"Total de posibles párrafos reales en Sinonimos: {parrafos.shape[0]}")

# Ver ejemplos
parrafos[["Palabra", "Sinonimos"]].head(50)
```

```python
# 1. Definir la condición de búsqueda
# Buscamos las filas que cumplan AMBAS condiciones:
# - Que la categoría ya limpia sea 'sust.'
# - Que el sinónimo (aún sin limpiar del todo) empiece con 'p.'
condicion = (df['Categoria_limpia'] == 'sust.') & (df['Sinonimos'].str.startswith('p.', na=False))

# 2. Modificar la columna 'Categoria_limpia' para esas filas
# Usando .loc, seleccionamos las filas que cumplen la condición y la columna 'Categoria_limpia',
# y le asignamos el nuevo valor 'sust. p.'.
df.loc[condicion, 'Categoria_limpia'] = 'sust. p.'

# 3. Limpiar la columna 'Sinonimos' en esas mismas filas
# Nuevamente con .loc, seleccionamos las filas y la columna 'Sinonimos'.
# - .str.removeprefix('p.') quita el prefijo "p."
# - .str.strip() elimina cualquier espacio en blanco que pueda quedar al inicio.
df.loc[condicion, 'Sinonimos'] = df.loc[condicion, 'Sinonimos'].str.removeprefix('p.').str.strip()

# --- Verificación ---
print("¡Corrección aplicada con éxito! ✨\n")

# Creamos una lista de índices que sabemos que tenían el problema para verificar
indices_ejemplo = [47, 58, 62, 87, 123, 323, 420, 506]
df.loc[indices_ejemplo, ['Palabra', 'Categoria_limpia', 'Sinonimos']]
```

```python
import re
import pandas as pd

# Compilar el patrón para buscar palabras cortadas
pattern = re.compile(r'(\w+)-\s+(\w+)', flags=re.UNICODE)

# Columnas a revisar
cols_a_revisar = ['Sinonimos', 'Descripcion', 'Antonimos']
matches = []

# Primera parte: Detectar las palabras cortadas
for idx, row in df.iterrows():
    for col in cols_a_revisar:
        texto = row.get(col)
        if pd.isna(texto):
            continue
        s = str(texto)
        for m in pattern.finditer(s):
            match_text = m.group(0)
            reconstruido = m.group(1) + m.group(2)
            contexto = s[max(0, m.start()-40): m.end()+40]
            matches.append({
                'index': idx,
                'Palabra': row.get('Palabra'),
                'columna': col,
                'match': match_text,
                'reconstruido': reconstruido,
                'contexto': contexto
            })

palabras_cortadas = pd.DataFrame(matches)

# Mostrar resumen de coincidencias encontradas
print(f"Coincidencias encontradas: {len(palabras_cortadas)}")
print(f"Filas afectadas: {palabras_cortadas['index'].nunique() if not palabras_cortadas.empty else 0}")

# Segunda parte: Corregir las palabras cortadas en el DataFrame original
if not palabras_cortadas.empty:
    print("\nEjemplos de correcciones a realizar:")
    display(palabras_cortadas.sort_values(['index', 'columna']).reset_index(drop=True))
    
    # Aplicar las correcciones al DataFrame original
    for _, match in palabras_cortadas.iterrows():
        idx = match['index']
        col = match['columna']
        texto_original = df.at[idx, col]
        
        if pd.isna(texto_original):
            continue
            
        # Reemplazar la palabra cortada con la versión reconstruida
        texto_corregido = pattern.sub(lambda m: m.group(1) + m.group(2), str(texto_original))
        df.at[idx, col] = texto_corregido
    
    print("\n¡Correcciones aplicadas! 🎉")
    
    # Mostrar algunos ejemplos de las correcciones realizadas
    indices_afectados = palabras_cortadas['index'].unique()[:5]  # Mostrar solo los primeros 5 ejemplos
    print("\nEjemplos de registros corregidos:")
    for idx in indices_afectados:
        print(f"\nRegistro {idx}:")
        for col in cols_a_revisar:
            if df.at[idx, col] is not None and not pd.isna(df.at[idx, col]):
                print(f"{col}: {df.at[idx, col]}")
else:
    print("\nNo se encontraron palabras cortadas con el patrón '\\w+-\\s+\\w'.")
```
